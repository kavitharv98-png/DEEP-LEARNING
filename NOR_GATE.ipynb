```python
import numpy as np

class Perceptron:
    def __init__(self, weights, bias):
        self.weights = np.array(weights)
        self.bias = bias

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights) + self.bias
        # Standard step activation function
        return 1 if summation >= 0 else 0

inputs = np.array([[0,0],[0,1],[1,0],[1, 1]])

# Weights [-1, -1] and Bias 0.0 create a NOR gate
nor_gate = Perceptron(weights=[-1, -1], bias=0.0)

print("NOR Gate Output:")
for x in inputs:
    print(f"{x} -> {nor_gate.predict(x)}")
```
