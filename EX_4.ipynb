```python
import numpy as np

class Perceptron:
    def __init__(self, weights, bias):
        self.weights = np.array(weights)
        self.bias = bias

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights) + self.bias
        return 1 if summation >= 0 else 0

# Standard 2-input truth table matrix
inputs = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Fixed exercise constraint
fixed_bias = -0.5

# 1. AND Gate Verification
and_perceptron = Perceptron(weights=[0.3, 0.3], bias=fixed_bias)
print("--- AND Gate Verification ---")
for x in inputs:
    print(f"{x} -> {and_perceptron.predict(x)}")

# 2. OR Gate Verification
or_perceptron = Perceptron(weights=[0.5, 0.5], bias=fixed_bias)
print("\n--- OR Gate Verification ---")
for x in inputs:
    print(f"{x} -> {or_perceptron.predict(x)}")

# 3. NAND Gate Demonstration (Showing the structural failure at [0,0])
nand_perceptron = Perceptron(weights=[-0.5, -0.5], bias=fixed_bias)
print("\n--- NAND Gate Attempt (Fails at [0,0]) ---")
for x in inputs:
    print(f"{x} -> {nand_perceptron.predict(x)}")
```
