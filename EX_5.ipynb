```python
import numpy as np

class Perceptron:
    def __init__(self, weights, bias):
        self.weights = np.array(weights)
        self.bias = bias

    # 1. Sigmoid Activation Function
    def activate_sigmoid(self, inputs):
        summation = np.dot(inputs, self.weights) + self.bias
        return 1 / (1 + np.exp(-summation))

    # 2. ReLU Activation Function
    def activate_relu(self, inputs):
        summation = np.dot(inputs, self.weights) + self.bias
        return max(0, summation)

inputs = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Using standard AND gate parameters
gate = Perceptron(weights=[1.0, 1.0], bias=-1.5)

print("--- Sigmoid Output ---")
for x in inputs:
    print(f"{x} -> {gate.activate_sigmoid(x):.4f}")

print("\n--- ReLU Output ---")
for x in inputs:
    print(f"{x} -> {gate.activate_relu(x):.4f}")
```
