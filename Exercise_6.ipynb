```python
# Perceptron Training for One Epoch

# Initial values
learning_rate = 0.2
weights = [0.3, 0.1]
bias = 0

# Training data (Change these values if needed)
inputs = [
    [1, 0],
    [0, 1],
    [1, 1],
    [0, 0]
]

targets = [1, 1, 1, 0]

print("Initial Weights:", weights)
print("Initial Bias:", bias)

# Train for one epoch
for i in range(len(inputs)):
    x = inputs[i]
    target = targets[i]

    # Net Input
    net = x[0] * weights[0] + x[1] * weights[1] + bias

    # Activation Function
    if net >= 0:
        output = 1
    else:
        output = 0

    # Error
    error = target - output

    # Update Weights
    weights[0] = weights[0] + learning_rate * error * x[0]
    weights[1] = weights[1] + learning_rate * error * x[1]

    # Update Bias
    bias = bias + learning_rate * error

    # Display Results
    print("\nTraining Sample", i + 1)
    print("Input:", x)
    print("Target:", target)
    print("Net Input:", net)
    print("Output:", output)
    print("Error:", error)
    print("Updated Weights:", weights)
    print("Updated Bias:", bias)

print("\nFinal Weights:", weights)
print("Final Bias:", bias)
```
